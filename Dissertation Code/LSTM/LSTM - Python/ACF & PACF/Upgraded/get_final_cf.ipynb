```python
# read xlsx file
import pandas as pd


folder = 'C:\\Users\\Omen\Documents\\GitHub\\Dissertation Code\\LSTM\LSTM - Python\\ACF & PACF\\Upgraded\\results_upgraded.xlsx'
df = pd.read_excel(folder)
df
```

```python
# add collumn named '(p-1,q),(p,q-1),(p-1,q-1)' to the dataframe
df['PoN'] = ''
df
```

```python
collumns = df.columns
collumns
```

```python
# get the value on row 0 and collumn 0
row = df.iloc[20,0]

# '(0,0)' to p=0 and q=0
row = row.replace('(','').replace(')','').split(',')
# convert string to int
row = [int(i) for i in row]
print(row)

pminus1 = 0
qminus1 = 0
pminus1q = 0

if row[0]-1 > 0:
    pminus1 = row[0]-1
if row[1]-1 > 0:
    qminus1 = row[1]-1

neighbours = [[pminus1, row[1]], [row[0], qminus1], [pminus1, qminus1]]
# remove duplicates
neighbours = list(set(map(tuple, neighbours)))

# if (row[0],row[1]) is in the list of neighbours, remove it
if tuple(row) in neighbours:
    neighbours.remove(tuple(row))

neighbours
```

```python
neighbours_sum = 0
for i in range(len(neighbours)):
    neighbour = str(neighbours[i])
    # remove the space
    neighbour = neighbour.replace(' ','')
    # get the index of neighour in the collumns
    for j in range(len(collumns)):
        #print(collumns[j], neighbour)
        if neighbour in collumns[j]:
            index = j
            break
    
    # get the value of the neighbour
    value = df.iloc[20,index]
    neighbours_sum += value
    
neighbours_sum
```

```python
paint = []

# iterate through the rows
for r in range(len(df)):
    row = df.iloc[r,0]
    row22 = row
    # '(0,0)' to p=0 and q=0
    row = row.replace('(','').replace(')','').split(',')
    # convert string to int
    row = [int(i) for i in row]
    
    pminus1 = 0
    qminus1 = 0
    pminus1q = 0

    if row[0]-1 > 0:
        pminus1 = row[0]-1
    if row[1]-1 > 0:
        qminus1 = row[1]-1

    neighbours = [[pminus1, row[1]], [row[0], qminus1], [pminus1, qminus1]]
    # remove duplicates
    neighbours = list(set(map(tuple, neighbours)))

    # if (row[0],row[1]) is in the list of neighbours, remove it
    if tuple(row) in neighbours:
        neighbours.remove(tuple(row))

    neighbours_sum = 0

    color_neighbours = []
    for i in range(len(neighbours)):
        neighbour = str(neighbours[i])
        # remove the space
        neighbour = neighbour.replace(' ','')
        color_neighbours.append(neighbour)
        # get the index of neighour in the collumns
        for j in range(len(collumns)):
            #print(collumns[j], neighbour)
            if neighbour in collumns[j]:
                index = j
                break

        # get the value of the neighbour
        value = df.iloc[r,index]
        neighbours_sum += value

    df.at[r, 'PoN'] = neighbours_sum
    paint.append([row22,color_neighbours])
```

```python
import openpyxl
from openpyxl.styles import PatternFill

# Load the workbook and select the active worksheet
file_path = 'results_upgraded_final.xlsx'
workbook = openpyxl.load_workbook(file_path)
sheet = workbook.active

# Define the red fill
red_fill = PatternFill(start_color="FFFF0000", end_color="FFFF0000", fill_type="solid")

# Get the dimensions of the worksheet
max_row = sheet.max_row
max_column = sheet.max_column

# Color the diagonals
for i in range(1, min(max_row, max_column) + 1):
    sheet.cell(row=i, column=i).fill = red_fill
    #sheet.cell(row=i, column=max_column - i + 1).fill = red_fill

# Save the workbook
workbook.save('results_upgraded_final.xlsx')
```

```python
import openpyxl
from openpyxl.styles import PatternFill
import pandas as pd

# Load the Excel workbook
file_path = 'results_upgraded_final.xlsx'
wb = openpyxl.load_workbook(file_path)
ws = wb.active

# Load the DataFrame
df = pd.read_excel(file_path)

for i in range(len(paint)):
    row_name = paint[i][0]
    # Strip leading and trailing whitespaces from column names
    df.columns = df.columns.str.strip()

    for j in range(len(paint[i][1])):
        column_name = paint[i][1][j]
        # color the row,collumn cell

        # Find the position of the row and column
        row_index = df.index[df['(p,q)'] == row_name].tolist()[0] + 2  # Add 2 to match Excel row index
        column_index = df.columns.get_loc(column_name) + 1  # Add 1 to match Excel column index

        # Get the cell value
        cell_value = df.loc[df['(p,q)'] == row_name, column_name].iloc[0]

        # Check if the cell value is positive float or integer
        if isinstance(cell_value, float) and cell_value > 0:
            # Set the fill color
            yellow_fill = PatternFill(start_color="FFFF00", end_color="FFFF00", fill_type="solid")

            # Apply the fill color to the cell
            cell_to_color = ws.cell(row=row_index, column=column_index)
            cell_to_color.fill = yellow_fill


# Save the workbook
wb.save("results_upgraded_final.xlsx")
```
